### Audio classification. 

```python
from datasets import load_dataset
from datasets import Audio
from transformers import pipeline
from IPython.display import Audio

minds = load_dataset("PolyAI/minds14", name="en-AU", split="train")
minds = minds.cast_column("audio", Audio(sampling_rate=16_000))

classifier = pipeline(
    "audio-classification",
    model="anton-l/xtreme_s_xlsr_300m_minds14",
    trust_remote_code=True
)
```

```python
example = minds[0]
classifier(example["audio"]["array"])
```

```python
id2label = minds.features["intent_class"].int2str
id2label(example["intent_class"s])
```

### Audio speech recognition. 

```python
from transformers import pipeline

asr = pipeline("automatic-speech-recognition")
example = minds[0]
asr(example["audio"]["array"])
```

```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
text = " WaveNet, a deep neural network for generating raw audio waveforms. The model is fully probabilistic and autoregressive, with the predictive distribution for each audio sample conditioned on all previous ones; nonetheless we show that it can be efficiently trained on data with tens of thousands of samples per second of audio."

output = pipe(text)
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
