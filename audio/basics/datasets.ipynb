```python
from datasets import load_dataset

minds = load_dataset("PolyAI/minds14", name="en-GB", split="train", trust_remote_code=True)
minds
```

```python
import IPython
print (minds[0]['english_transcription'])
IPython.display.Audio(minds[0]['audio']['path'])
```

```python
import gradio as gr

id2label = minds.features["intent_class"].int2str
id2label(minds[0]["intent_class"])

def generate_audio():
    example = minds.shuffle()[0]
    audio = example["audio"]
    return (
        audio["sampling_rate"],
        audio["array"],
    ), id2label(example["intent_class"])


with gr.Blocks() as demo:
    with gr.Column():
        for _ in range(4):
            audio, label = generate_audio()
            output = gr.Audio(audio, label=label)

demo.launch(debug=True)
```
